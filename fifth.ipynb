```python
from datetime import datetime
from langchain.tools import tool



# first tool
@tool
def get_current_time():
    """Returns the current system time."""

    return datetime.now().strftime("%H:%M:%S")
```

```python
# second tool
@tool
def calculator(a: float, b: float):
    """
    Adds two numbers.
    """
    return a + b
```

```python
from langchain.chat_models import init_chat_model

def get_model():
    return init_chat_model(model="groq:llama-3.3-70b-versatile")

model = get_model()
```

```python
tool_registry = {
    "get_current_time":get_current_time,
    "calculator":calculator
}

```

```python
model_with_tools = model.bind_tools(list(tool_registry.values()))
```

```python
response = model_with_tools.invoke("what time it is?")
response.tool_calls
```

```python
response = model_with_tools.invoke(
    "What is 45 plus 89?"
)
tool_call = response.tool_calls[0]
tool_name = tool_call["name"]
tool_args = tool_call["args"]

tool = tool_registry[tool_name]


response = tool.invoke(tool_args)

response

```
